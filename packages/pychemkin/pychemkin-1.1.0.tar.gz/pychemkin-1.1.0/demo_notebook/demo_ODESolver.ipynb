```python
import sys
sys.path.append('../')
from pychemkin import ChemSolver, chemkin
import numpy as np
```

```python
chem = chemkin("demo_xml/rxns.xml")
# initial concentration
x_init = [1]*6
T = 1000
# time series you want
ts = 1.e-7
# set to (min val in ts, max val in ts)
t_max = 5.e-5
cs = ChemSolver(chem).solve(x_init, T, t_max, ts)
df = cs.to_df()
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize = (10,5))
plt.plot(df['t'],df['H2-Concentration'])
plt.plot(df['t'],df['H2O-Concentration'])
plt.plot(df['t'],df['O-Concentration'])
plt.plot(df['t'],df['H-Concentration'])
plt.plot(df['t'],df['OH-Concentration'])
plt.plot(df['t'],df['O2-Concentration'])
plt.legend()
df.tail(20)
```
